```python
from clustering import run, read
import os
import matplotlib.pyplot as plt
from style_feature import Style_Feature
%matplotlib inline
```

```python
def main(mode, data_path, cluster_folder, n_cluster_range, f_lab, f_gist, cluster_model, processed_model):
    style_obj = Style_Feature(f_lab, f_gist, processed_model)    
    features, img_path = read(data_path, style_obj)
    for n_cluster in n_cluster_range:
        folder = '%s/res_%d' % (cluster_folder, n_cluster)
        os.makedirs(folder)
        run(mode, features, img_path, folder, n_cluster, cluster_model)
```
